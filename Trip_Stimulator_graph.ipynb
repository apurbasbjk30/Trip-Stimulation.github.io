```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate data: speeds for each user for 5 trips
user_names = ["User 1", "User 2", "User 3", "User 4", "User 5"]

# Each user has 5 trips with different speeds (in km/h)
user_trip_speeds = [
    [60, 55, 70, 65, 75],  # User 1's speeds
    [50, 52, 55, 58, 60],  # User 2's speeds
    [65, 70, 75, 80, 85],  # User 3's speeds
    [55],                  # User 4's speeds
    [60]                   # User 5's speeds
]

# Calculate the average speed for each user
user_avg_speeds = [np.mean(trip_speeds) for trip_speeds in user_trip_speeds]

# Create a bar chart to visualize the average speeds for each user
plt.figure(figsize=(10, 6))
plt.bar(user_names, user_avg_speeds, color='blue', label="Average Speed for Each User")

# Add labels and title
plt.xlabel("User")
plt.ylabel("Average Speed (km/h)")
plt.title("Average Speed for Each User's 5 Trips")
plt.legend()

# Show the plot
plt.show()
```
