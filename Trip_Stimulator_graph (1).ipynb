```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate data: speeds for each trip
Trip_names = ["Trip 1", "Trip 2", "Trip 3", "Trip 4", "Trip 5"]

# Speeds for each trip
user_trip_speeds = [
    [60, 55, 70, 65, 75],  # Trip 1's speeds
    [50, 52, 55, 58, 60],  # Trip 2's speeds
    [65, 70, 75, 80, 85],  # Trip 3's speeds
    [55],                  # Trip 4's speeds
    [60]                   # Trip 5's speeds
]

# Calculate the average speed for each trip
user_avg_speeds = [np.mean(trip_speeds) for trip_speeds in user_trip_speeds]

# Calculate the best average speed (maximum of the average speeds)
best_avg_speed = max(user_avg_speeds)

# Create a bar chart to visualize the average speeds for each trip
plt.figure(figsize=(10, 6))
plt.bar(Trip_names, user_avg_speeds, color='blue', label="Average Speed for Each Trip")

# Add labels and title
plt.xlabel("Trip")
plt.ylabel("Average Speed (km/h)")
plt.title("Average Speed for Each Trip")

# Add a horizontal line for the best average speed
plt.axhline(y=best_avg_speed, color='red', linestyle='--', label=f"Best Average Speed: {best_avg_speed} km/h")

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
